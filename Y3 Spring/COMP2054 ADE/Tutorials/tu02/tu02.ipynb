```python
import math
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
def plot_oh(f, g, c, n_0, min_x, max_x):
    xs = np.arange(min_x, max_x, 1, dtype='int64')  # create a list with min, max and step = 1
    fn_ys = f(xs)
    gn_ys = c*g(xs)
    plt.axvline(x = n_0, color = 'k', linestyle='--')
    plt.scatter(x=xs, y=fn_ys, marker='x')
    plt.scatter(x=xs, y=gn_ys, marker='.')
    plt.xlabel('n')
    plt.ylabel('f(n)')
    plt.legend(['n_0 = '+str(n_0),'f(n)', 'cg(n)'])
```

## Big-Oh Definition
Given positive functions $f(n)$ and $g(n)$, we can say that $f(n) \text{ is } O(g(n))$ if and only if there exists positive constants $c$ and $n_0$ such that $f(n) \leq c \cdot g(n), \forall n \geq n_0$.

### Q1. Prove that $5$ is $O(1)$

We wanna prove, 
$$5 \le 1 \cdot c, \ \forall n \ge n_0$$

Take $c = 5$, we have
$$5 \le 5, \ \forall n \ge n_0$$

This is trivially true for all positive $n$. Here we take $n_0 = 1$, then we have
$$5 \le 5, \ \forall n \ge 1$$ 

Therefore $5$ is $O(1)$ using $c = 1, n_0 = 1$

### Sanity Check

```python
f = lambda n: np.full(n.shape, 5)
g = lambda n: np.full(n.shape, 1)
c = 5
n_0 = 1

plot_oh(f, g, c, n_0, 1, 100)
```

### Q12. Given that $f(n) =\text{ IF }even(n)\text{ THEN }n+3\text{ ELSE }n^2 + 5$ state the Big-Oh Behaviour and prove it from the definition

We wanna prove $f(n)$ is $O(n^2)$, let's consider two cases:

When $n$ is **even**
$$n + 3 \le c \cdot n^2, \ \forall n \ge n_0$$

Simplify the inequation by divided by $n^2$, we have
$$1/n + 3/n^2 \le c, \ \forall n \ge n_0$$

Take $c = 2$, we have
$$ 1/n + 3/n^2 \le 2 \Rightarrow n \ge 1.5, \ \forall n \ge n_0 $$

Take $n_0 = 2$, we have
$$n \ge 1.5, \ \forall n \ge 2 $$

---
When $n$ is **not even**

$$ n^2 + 5 \le c \cdot n^2, \ \forall n \ge n_0$$

Simplify the inequation by divided by $n^2$, we have
$$ 1 + 5/n^2 \le c, \ \forall n \ge n_0 $$

Take $c = 2$, we have
$$ 1 + 5/n^2 \le 2 \Rightarrow n \ge \sqrt5, \ \forall n \ge n_0 $$

Take $n_0 = 3$, then we have
$$ n \ge \sqrt5, \ \forall n \ge n_0 $$

In total, $f(n)$ is $O(n^2)$ using $c = 2, n_0 = 3$

```python
c = 2.25
n_0 = 2
f = lambda n: np.where(n % 2 == 0, n + 3, n**2 + 5)
g = lambda n: n**2

plot_oh(f, g, c, n_0, 0, 50)
#plt.rcParams["figure.figsize"] = (4,4)
```

$$2^n \ge n, \ \forall n \ge 1 $$

Therefore we have $2 ^ n + n$ is $O(2 ^ n)$ using $c = 2, n_0 = 1$
